```python
import os
import numpy as np
from PIL import Image

# Define data directories (adjust paths if necessary)
train_dir = 'Base de dados/train'

#as pastas train e test1 devem ser baixadas do site abaixo:
#https://www.kaggle.com/c/dogs-vs-cats
```

```python
size = 96

# Define the function to preprocess an image
def preprocess_image(image_path):
    image = Image.open(image_path)
    image = image.convert('L')  # Convert image to grayscale
    image = image.resize((size, size))  
    image = np.array(image)
    image = image.astype('float32')
    image /= 255.0
    image = (image - 0.5) * 2.0  # Apply contrast enhancement
    return image
```

```python
# Load training and test data
train_data = []

for class_name in os.listdir(train_dir):
    for image_file in os.listdir(os.path.join(train_dir, class_name)):
        image_path = os.path.join(train_dir, class_name, image_file)
        print(image_path)
        image = preprocess_image(image_path)
        label = class_name  # Extract label from class name
        train_data.append((image, label))  # Append as a tuple (image, label)
```

```python

# Separate features and labels from the list of tuples
train_images, train_labels = zip(*train_data)

```

```python
# Define a dictionary to map string labels to integers
label_map = {
    "cat": 0,
    "dog": 1,
     # Add the missing label mapping
}

# Convert string labels to integers using the label_map
train_labels = np.array(list(map(lambda x: label_map[x], train_labels)))
```

```python
# Reshape images (if necessary)
input_shape = (size, size, 1)  # Assuming grayscale images
train_images = np.array([image.reshape(input_shape) for image in train_images])
```

```python
# Import libraries

import matplotlib.pyplot as plt
plt.imshow(train_images[150], cmap='gray')
```

```python
# caso a rede j esteja salva, v at a ltima clula para carregar a rede salva
import tensorflow as tf

model = tf.keras.models.Sequential()

model.add(tf.keras.layers.Conv2D(32, (3,3), activation="relu", input_shape=input_shape))
model.add(tf.keras.layers.MaxPooling2D(2,2))
model.add(tf.keras.layers.Conv2D(64, (3,3), activation="relu"))
model.add(tf.keras.layers.MaxPooling2D(2,2))
model.add(tf.keras.layers.Conv2D(128, (3,3), activation="relu"))
model.add(tf.keras.layers.MaxPooling2D(2,2))
model.add(tf.keras.layers.Conv2D(128, (3,3), activation="relu"))
model.add(tf.keras.layers.MaxPooling2D(2,2))

model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dropout(0.5)) 
model.add(tf.keras.layers.Dense(units=512, activation='relu'))
model.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid')) 
model.summary()
```

```python
# Compile the model
model.compile(loss='binary_crossentropy', optimizer='rmsprop', metrics=['acc'])
```

```python
# Define the early stopping callback
stopper = tf.keras.callbacks.EarlyStopping(
    monitor="loss", patience=10, verbose=2
)
```

```python
from tensorflow.keras.callbacks import ReduceLROnPlateau

# Define the learning rate reduction callback
reduce_lr = ReduceLROnPlateau(monitor='val_loss', patience=3, factor=0.2, verbose=1)
```

```python

# Initialize lists to store history
epoch_hist = model.fit(train_images, train_labels, epochs = 100, batch_size = 32, 
                       callbacks = [stopper,reduce_lr], validation_split = 0.20)
```

```python
plt.plot(epoch_hist.history['loss'])
plt.plot(epoch_hist.history['val_loss'])
plt.title('Model loss progress during training')
plt.xlabel('Epochs')
plt.ylabel('loss')
plt.legend(['traint loss', 'validate loss']);
```

```python
plt.plot(epoch_hist.history['acc'])
plt.plot(epoch_hist.history['val_acc'])
plt.title('Model accuracy progress during training')
plt.xlabel('Epochs')
plt.ylabel('accuracy')
plt.legend(['train accuracy', 'validate accuracy']);
```

```python
def class_name(pred):
  if pred<0.5:
    return "gato"
  else:
    return "cachorro"
```
